# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

---

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display the first 5 values of y
y.values[:5]

```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns="credit_ranking").copy()

# Review the features DataFrame
loans_df.head()
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
X_scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_scaler.fit(X_train)

# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
input_nodes = len(X.columns)

# Review the number of features
input_nodes
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes1 = 6

# Define the number of hidden nodes for the second hidden layer
hidden_nodes2 = 3

# Define the number of neurons in the output layer
hidden_nodes3 = 1
```

```python
# Create the Sequential model instance
nn_model = tf.keras.models.Sequential()

# Add the first hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes1, activation="relu", input_dim=input_nodes))

# Add the second hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes2, activation="relu", input_dim=input_nodes))

# Add the output layer to the model specifying the number of output neurons and activation function
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes3, activation="sigmoid"))
```

```python
# Display the Sequential model summary
nn_model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 50 epochs and the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled, y_test, verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Step 4: Save and export your model to a keras file, and name the file `student_loans.keras`.


```python
# Set the model's file path
model_path = 'student_loans.keras'

# Export your model to a keras file
nn_model.save(model_path)
```

### Step 1: Reload your saved model.

```python
from tensorflow.keras.models import load_model

# Set the model's file path
model_path = Path("student_loans.keras")

# Load the model to a new object
loaded_model = load_model(model_path)
```

### Step 2: Make predictions on the testing data and save the predictions to a DataFrame.

```python
# Make predictions with the test data
predictions = nn_model.predict(X_test_scaled,verbose=2)
predictions[:5]
```

```python
# Save the predictions to a DataFrame and round the predictions to binary results
predictions_rounded = [round(prediction[0],0) for prediction in predictions]
predictions_df = pd.DataFrame(
    {'predictions':predictions_rounded})

# Review the DataFrame
predictions_df
```

### Step 4: Display a classification report with the y test data and predictions

```python
# Print the classification report with the y test data and predictions
print(classification_report(y_test, predictions_rounded))
```

**1. Describe the data that you would need to collect to build a recommendation system to recommend student loan options for students. Explain why this data would be relevant and appropriate.**


If I'm recommending student loan products to students, it's important to know how much money they need, what other student loans they have, any other debt outstanding, and information about their short and medium term ability to pay. It will also be necessary to know the repayment terms of any loan products, when repayment begins, potential refinance terms, rate terms (fixed vs. variable, and if variable those terms), the lender's basic information, and any other information about the loan product itself.

It's necessary to know this information, because not all products will work for every student. For example, a student with low ability to repay in the short term should not take a high interest rate loan that starts repayment immediately after graduation with no flexibility in rate or repayment terms. The more information there is about a person's ability to pay, their needs, and what the product offers (and at what cost), the more likely a person will be to be recommended a loan product that works for them.


**2. Based on the data you chose to use in this recommendation system, would your model be using collaborative filtering, content-based filtering, or context-based filtering? Justify why the data you selected would be suitable for your choice of filtering method.**

I would argue it would use collaborative filtering. It might not be the case that the user has a long history of selecting student loan products, which would rule out content and context-based filtering. Those types of filtering take into account someone's past decisions (context-based includes circumstances around the decision as well). In the case of collaborative filtering, profiles of similar students could be created, and then used to generate recommendations that worked well for those individuals. There is less reliance on the past history of selecting loan products with collaborative filtering.


**3. Describe two real-world challenges that you would take into consideration while building a recommendation system for student loans. Explain why these challenges would be of concern for a student loan recommendation system.**


1. One main challenge would be trying to find the right financial information (features) that statistically relate to a student's ability to pay. Students often have little to no credit history, and it's very difficult to predict what their future earning prospects look like. There may be adverse economic or labor market conditions that have nothing to do with the student, but could dramatically affect their future earnings and ability to pay. Trying to find a reliable measure for their ability to repay a loan will affect what type of loan they can be offered. You don't want to recommend loans that would be too costly or unforgiving in cases of financial hardship.

2. Another challenge would be getting enough student data in general to create profiles to use as a base for the collaborative filtering. This would require getting consent from the students for their financial information and loan selection, coming up with secure infrastructure that would protect this data, and testing to make sure there is enough reliable information to use for a viable recommendation engine. This presents both marketing and technical challenges, as you would have to come up with a way to convince people to give their information to the project.
